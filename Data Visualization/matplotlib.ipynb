### Data Visualization: 
#### Representing the data in the form of pictures or graphs is called "Data Visualization".

```python
# We will use "pyplot" submodule of "matplotlib" module.
import matplotlib.pyplot as plt # for drawing the graph.
import pandas as pd # for creating the dataframe.
```

```python
df = pd.read_csv("empdata.csv")
df
```

### Bar Graph
#### Bar graph is used to represent categorical data with rectangular bars.

```python
# Program 1: A Python Program to display employee id numbers on x-axis and their salaries on Y-axis in the form a bar graph.

x = df["empid"]
y = df["salary"]

plt.bar(x, y, label="Employee Data", color="red") # this is what creates a bar chart

plt.xlabel("Employee id's") # print label in x-axis.
plt.ylabel("Employee salaries") # print label in y-axis.

plt.title("AR Inc") # print title

plt.legend() # this displays the legend colors and labels of data.

plt.show()
```

```python
# Program 2: A program to dispaly employee id numbers on X-axis and their salaries on Y-axis in the form of a 
# bar graph for two departments of a company.

# take employee ids and salaries for sales dept.
x = [1001, 1003, 1006, 1007, 1009, 1011]
y = [10000, 23000.50, 18000.33, 16500.50, 12000.75, 9999.99]

# take employee ids and salaries for production dept.
x1 = [1002, 1004, 1005, 1008, 1010, 1012]
y1 = [15000.50, 25000.75, 17500.25, 16000.50, 13000.99, 10500.00]

# create bar graphs
plt.bar(x, y, label="Sales Dept", color="red")
plt.bar(x1, y1, label="Production Dept", color="blue")

# set labels and title
plt.xlabel("Employee id's")
plt.ylabel("Employee salaries")
plt.title("AR Inc")
plt.legend()

# display the graph
plt.show()
```

```python
# Ware House data
wh_df = pd.read_csv("warehouse_data.csv")
```

```python
# Program 4: Group the data by year followed by daily_dispatch_count then plot
# the bar chart by year x-axis and daily_dispatch_count on y-axis.
wh_df_g = wh_df.groupby("year").agg({"daily_dispatch_count":"sum"}).reset_index()
plt.bar(wh_df_g["year"], wh_df_g["daily_dispatch_count"])
plt.xlabel("Year of Dispatch")
plt.ylabel("Dispatch count")
plt.title("Warehouse Data")
```

### Historgram
#### Histogram shows distributions of values. Histogram is similar to bar graph but it is useful to show values grouped in bins or intervals.

```python
# Program 3: A program to display a histogram showing the number of employees in specific age groups.
emp_ages = [22, 45, 30, 59, 58, 56, 57, 45, 43, 43, 50, 40, 34, 33, 25, 19]
bins = [0, 10, 20, 30, 40, 50, 60]

plt.hist(emp_ages, bins, histtype="bar", rwidth=0.8, color="cyan")

plt.xlabel("Employee Ages")
plt.ylabel("Number of Employees")
plt.title("AR Inc")
plt.legend()

plt.show()
```

### Pie Chart
#### Pie chart is used to represent data in the form of a circular graph. The circle is divided into sectors or slices that represent a proportion of the whole.

```python
# Program 4: A program tp display a pie chart showing the percentage of employees in different departments of a company.
depts = ["Sales", "Production", "HR", "IT"]
slices = [30, 40, 10, 20]

col = ["red", "blue", "green", "cyan"]

plt.pie(slices, labels=depts, colors=col, startangle=90, shadow=True, explode=(0, 0.2, 0, 0), autopct="%.1f%%")

plt.title("AR Inc")
plt.legend()

plt.show()
```

### Line Graph
#### Line graph is used to represent data points connected by straight lines. It is useful to show trends over time.

```python
# Program 5: A Program to create a line graph to show the profits of a company in various years.
years = [2015, 2016, 2017, 2018, 2019, 2020]
sales = [1000, 1200, 1500, 1700, 1600, 1800]
plt.plot(years, sales, marker='o', linestyle='-', color='b')

plt.title("Sales Data")
plt.xlabel("Years")
plt.ylabel("Sales")

plt.show()
```

```python
# Multiline ploting
# Aggregate data
wh_df_a = wh_df[wh_df["Product_Type"] == "Type_A"].groupby("month").agg({"daily_dispatch_count": "sum"}).reset_index()
wh_df_b = wh_df[wh_df["Product_Type"] == "Type_B"].groupby("month").agg({"daily_dispatch_count": "sum"}).reset_index()

# Ensure the months align (e.g., same order, same length)
# You can do an outer merge to align months properly
merged_df = wh_df_a.merge(wh_df_b, on="month", how="outer", suffixes=("_a", "_b")).fillna(0)

# Sort by month if needed
merged_df = merged_df.sort_values("month")

# Define x and y for plotting
x = merged_df["month"]
y1 = merged_df["daily_dispatch_count_a"]
y2 = merged_df["daily_dispatch_count_b"]

# Plot area between curves
plt.fill_between(x, y1, y2, color="lightblue", alpha=0.5)

# Plot lines
plt.plot(x, y1, color="#990000", marker="*", label="Type A")
plt.plot(x, y2, color="#6495ED", marker="*", label="Type B")

# Labels and grid
plt.grid(True)
plt.xlabel("Month")
plt.ylabel("Daily Dispatch Count")
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Program 6: To predict the number of advertisements based on targeted sales using the Linear Regression machine learning model.
# To predict Advt cost depending on targeted Sales
# using Linear Regression machine learing model
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data into dataframe
df = pd.read_csv("advt.csv")

# take the sales column as input data (x)
# take TV-ads column as target data (y)
x = df.iloc[:, 1:2]
y = df.iloc[:, 2]

# plot a scatter plot to know the relation b/w x and y
plt.scatter(x,y, color="red")
plt.xlabel("Sales ($)")
plt.ylabel("Number of TV Ads")
plt.show()
```

```python
# above scatter plot shows that the points can be shown on a straight line. Hence we can use Linear Regression model.
# create Linear Regression model
model = LinearRegression()
model.fit(x, y) # fitting means training the model with data

#find the accuracy of the model
# it is giving 93% accuracy

model.score(x, y) # output: 0.9338656808250994

# for all x values, the model can predict its own y values.
predicted_y = model.predict(x)

# plot a scatter plot again with predicted y values
# straight line indicates output from linear regression model.
plt.xlabel("Sales ($)")
plt.ylabel("Number of TV Ads")
plt.scatter(x, y, color="red", marker='+')
plt.plot(x, predicted_y, color="blue") #pyplot is used to plot the line graph.

# find how many ads to be given to reach sales target of 20000$
model.predict([[20000]]) # output: array([23.73632413])
# hence to reach sales target of $20000, 24 ads are to be given.
```
